```
from google.colab import drive
drive.mount('/content/drive',force_remount=True)
import numpy as np
import matplotlib.pyplot as plt
import pandas
df=pandas.read_csv('/content/drive/My Drive/Classroom/datasample.csv')
x = df['e_prevtx_eligible'].fillna(value=0)
y = df['e_prevtx_eligible_lo'].fillna(value=0)
x_mean = x.mean(axis=0)
y_mean = y.mean(axis=0)
n = len(x)
numerator = 0
denominator = 0
for i in range(n):
  numerator += (x[i] - x_mean) * (y[i] - y_mean)
  denominator += (x[i] - x_mean) **2
theta_1 = numerator / denominator
theta_0 = y_mean - (theta_1 * x_mean)
print('theta_1:',theta_1,'theta_0:',theta_0)
y = theta_0 + theta_1 * x
plt.plot(x, y, color='black')
plt.scatter(x, y, color='red')
plt.xlabel('e_prevtx_eligible')
plt.ylabel('e_prevtx_eligible_lo')
plt.show()
```
